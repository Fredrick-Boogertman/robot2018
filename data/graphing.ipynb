```python
data = read_csv('drivetrain.csv')

ts = data['time']
vels_l = data['enc_vel_l']
vels_r = data['enc_vel_r']
desired = data['computed_velocity']

plt.plot(ts, vels_l)
plt.xlabel('time (s)')
plt.ylabel('velocity (enc/100ms)')
plt.plot(ts, vels_r)
plt.plot(ts, desired)
plt.legend(["Velocity left", "Velocity right", "Velocity desired"])

plt.show()
```
